```python
# Filtering data records
df[df['Age'] > 25]
```

```python
# Filter age is greater then 25 and Annual Income
# use and as -> & and or as -> &
df[(df['Age'] > 25) & (df['Annual Income (k$)']>100)]
```

```python
df[(df['Age'] > 25) & (df['Annual Income (k$)']>100)].shape
```

```python
# Groping
df.head()
```

```python
# how many male and female customers
df.groupby('Gender')['CustomerID'].count()
```

```python
# For each Gender tell me total customers (count) and total Income (sum)
df.groupby('Gender').agg({'CustomerID':'count','Annual Income (k$)':'sum'})
```

```python
df.columns
```

```python
df.groupby('Gender').agg(TotalCustomers = ('CustomerID','count'), TotalIncome =('Annual Income (k$)','sum'))
```
