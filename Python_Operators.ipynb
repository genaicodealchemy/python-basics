```python
cls = 12
age = 17

print(cls >= 12)
print(age >= 18)
```

```python
# AND Operator
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```
