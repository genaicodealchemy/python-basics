Break and Continue statements


````python
val = int(input("Enter a number: "))
num = 1
while(num <= 10):
  if(num == val):
    break
  else:
    print(num)
  num = num + 1
````

````python
val = int(input("Enter a number: "))
num = 1
while(num <= 10):
  if(num == val):
    continue
  else:
    print(num)
  num = num + 1
````

````python
val = 3
for i in range(1,11):
  if(i == val):
    continue
  else:
    print(i)
````
