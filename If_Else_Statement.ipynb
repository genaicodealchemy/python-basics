```python
# if the number [1,10] square it , [11-20] cube it, else print the number as it is
num = int(input("Enter the number"))

if((num >= 1) and (num <=10)):
  res = num * num
  print("The result is ", res)
elif((num >=11) and (num<=20)):
  res = num ** 3
  print("The result is ", res)
else:
  print("The result is ", num)
```
